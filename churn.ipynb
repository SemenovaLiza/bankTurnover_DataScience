```python
def info_(df, sample_n):
    df_sample = df.sample(sample_n)
    df_info = df.info()
    print(f"{sample_n} SIZED SAMPLE")
    print()
    print(df_sample)
    print()
    print(df_info)
```

```python
def run_function(dfs, function, *fun_args):
    for name, features in dfs.items():
        print(50*"-"+name.upper()+50*"-")
        print(function(dfs[name], *fun_args))
run_function(dfs, info_, 5)
```

```python
def study_null(df, null_only=False):
    null_sum = df.isna().sum()
    df_null = pd.concat([df.dtypes, null_sum, null_sum / len(df) * 100], axis=1)
    df_null.columns = ["dtypes", "null_count", "missing_%"]
    if null_only:
        df_null = df_null[df_null['null_count'] > 0]
    df_null.sort_values(by='missing_%', ascending=False)
    return df_null
```

```python
run_function(dfs, study_null)
```

```python
def describe_full(df):
    df_describe = df.describe().T
    df_numeric = df._get_numeric_data()
    df_describe['dtypes'] = df_numeric.dtypes
    df_describe['missing_%'] = df_numeric.isna().sum() / len(df_numeric) * 100
    cardinality = df_numeric.apply(pd.Series.nunique)
    df_describe['Cardinality'] = cardinality
    df_describe['Skew'] = df_numeric.skew(axis=0, skipna=False)
    return df_describe
```

```python
run_function(dfs, describe_full)
```

```python
def convert_type(df):
    for col in df.select_dtypes('object').columns:
        if sorted(df[col].dropna().unique())==['No', 'Yes']:
            # df[col] = (df[col] == 'Yes').astype('int')
            df[col] = df[col].map({'No': 0, 'Yes': 1})
        elif sorted(df[col].dropna().unique())==['False', 'True']:
            df[col] = df[col].map({'False': 0, 'True': 1})
    return df

telecom_df = convert_type(telecom_df)
telecom_df
```

```python
def display_group_density_plot(df, groupby, on, palette=None, figsize=None, title='', ax=None):
    if palette is None:
        palette = sns.color_palette('Set2')
    if figsize is None:
        figsize = (10, 5)

    if not isinstance(df, pd.core.frame.DataFrame):
        raise ValueError('df must be a pandas DataFrame')
    if not groupby:
        raise ValueError('groupby parameter must be provided')
    elif groupby not in df.columns:
        raise ValueError(f'{groupby} column does not exist in the given DataFrame')
    if not on:
        raise ValueError('on parameter must be provided')
    elif on not in df.columns:
        raise ValueError(f'{on} column does not exist in the given DataFrame')

    if len(set(df[groupby])) > 10:
        groups = df[groupby].value_counts().index[:10]
    else:
        groups = set(df[groupby])

    if palette:
        palette = palette[:len(groups)]
    else:
        palette = sms.color_palette()[:len(groups)]

    if ax is None:
        fig = plt.figure(figsize=figsize)
        ax = fig.add_subplot(111)

    for value, color in zip(groups, palette):
        sns.kdeplot(df.loc[df[groupby] == value][on], \
                    shade=True, color=color, label=value, ax=ax)
    ax.legend(bbox_to_anchor=(1.04, 1), loc='upper left')

    if not title:
        title = f'Distribution of {on} per {groupby} group'

    ax.set_title(title, fontsize=16)
    ax.set_xlabel(on, fontsize=16)
    return ax    
```

```python
def display_count_plot(df, target):
    # split into categories and target
    features = list(set(df.columns) - set([target]))
    n_cols = 3
    n_rows = math.ceil(len(features) / n_cols)

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5 * n_rows))
    axes = axes.flatten()

    for ax, col in zip(axes, features):
        ct = pd.crosstab(df[col], df[target], normalize='index')
        ct.plot(kind='bar', stacked=True, ax=ax, legend=False)
        ax.set_title(col)
        ax.set_ylabel('Proportion')
        ax.tick_params(axis='x', rotation=45)

    # remove empty subplots
    for i in range(len(features), len(axes)):
        fig.delaxes(axes[i])

    fig.legend(
        title=target,
        bbox_to_anchor=(1.02, 1),
        loc='upper left'
    )

    plt.tight_layout()
    return fig
```

```python
def display_box_plot(df, x_, y_, ax=None):
    sns.boxplot(x=x_, y=y_, hue=x_, data=df, ax=ax)
    ax.legend(bbox_to_anchor=(1.04, 1), loc='upper right')
    ax.set_xlabel('')
    ax.set_xticks([])
    ax.set_title(x_)
    return ax
```
